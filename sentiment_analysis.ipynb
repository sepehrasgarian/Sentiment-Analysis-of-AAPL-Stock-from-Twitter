```python
# NLTK VADER for sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import pandas as pd 
```

```python
def preprocess(data):
    import re
    import string 
    data['text']=data['text'].str.replace(" ", " ")
    data['text'] = data['text'].map(lambda x: re.sub('https|com|tweet|twitter', '', x))



    data['paper_text_processed'] = data['text'].map(lambda x: re.sub('[,\.!?]', '', x))


    # Convert the titles to lowercase
    data['paper_text_processed'] = data['paper_text_processed'].map(lambda x: x.lower())
    data['paper_text_processed'] = data['paper_text_processed'].map(lambda x: re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', x)) # remove URLs
    data['paper_text_processed'] = data['paper_text_processed'].map(lambda x: re.sub('@[^\s]+', 'AT_USER', x))
    data['paper_text_processed'] = data['paper_text_processed'].map(lambda x: re.sub(r'#([^\s]+)', r'\1', x))
    def remove_punct(text):
        text  = "".join([char for char in text if char not in string.punctuation])
        text = re.sub('[0-9]+', '', text)
        return text

    data['paper_text_processed'] = data['paper_text_processed'].apply(lambda x: remove_punct(x))
    def tokenization(text):
        text = re.split('\W+', text)
        return text
    data['paper_text_processed'] = data['paper_text_processed'].replace(r'\\n',  ' ', regex=True).replace(r'\r',  ' ', regex=True).replace(r'\r\n',  ' ', regex=True).replace(r'\n',  ' ', regex=True)


    stopword = nltk.corpus.stopwords.words('english')
    def remove_stopwords(text):
        text = [word for word in text if word not in stopword]
        return text
    #data['paper_text_processed'] = data['paper_text_processed'].apply(lambda x: remove_stopwords(x)) 
    data['paper_text_processed'] = data['paper_text_processed'].map(lambda x: re.sub('^https?:\/\/.*[\r\n]*', '', x, flags=re.MULTILINE))   
    return data['paper_text_processed']
```
